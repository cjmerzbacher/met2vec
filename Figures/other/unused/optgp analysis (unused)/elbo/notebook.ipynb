```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import colormaps as cmaps
import distinctipy as dpy
```

```python
def load_nr16(k):
    return pd.read_csv(f"./k{k}_nr16_ns5k.csv")

ks = [250, 500, 1000, 2500, 5000]

nr16_dfs = {
    k : load_nr16(k)
    for k in ks
}

nr16_dfs[250]
```

```python
colors = dpy.get_colors(len(nr16_dfs), pastel_factor=0.8)

for i, (k, k_df) in enumerate(nr16_dfs.items()):
    print(k, k_df['pre'].max())
    plt.plot(k_df["k"], k_df['pre'], label=k, c=colors[i])

plt.ylabel("Inertia")
plt.xlabel("$k$ (Kmeans)")

plt.legend(title="$k$ (Optgp)")
plt.title("KMeans Clustering Inertia")
```
