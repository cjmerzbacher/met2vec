```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import distinctipy as dpy
```

```python
hc_pret_df = pd.read_csv("hc_pre_tsne.csv")
hc_embt_df = pd.read_csv("hc_emb_tsne.csv")
hc_rect_df = pd.read_csv("hc_rec_tsne.csv")
```

```python
cell_types = list(hc_pret_df["label"].unique())
cell_colors = np.array(dpy.get_colors(len(cell_types), pastel_factor=0.8))

def tshow(df : pd.DataFrame, ax, label=""):
    colors = cell_colors[df["label_n"]]

    ax.scatter(df["tsne0"], df["tsne1"], s=1, c=colors)
    ax.set_title(label)

def tshow_(df : pd.DataFrame, ax, label=""):
    colors = cell_colors[df["label_n"]]

    ax.scatter(df["tsne0"], df["tsne1"], s=1, c=colors)
    ax.set_title(label)
```

```python

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18,6))

hc_pret_df["label_n"] = hc_pret_df["label"].map(cell_types.index)
hc_embt_df["label_n"] = hc_embt_df["label"].map(cell_types.index)
hc_rect_df["label_n"] = hc_rect_df["label"].map(cell_types.index)

tshow(hc_pret_df, ax1, "Pre")
tshow(hc_embt_df, ax2, "Emb")
tshow(hc_rect_df, ax3, "Rec")

fig.savefig("After.png")
```
