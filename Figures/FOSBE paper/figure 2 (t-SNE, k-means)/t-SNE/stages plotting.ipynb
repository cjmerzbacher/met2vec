```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import distinctipy as dpy
```

# Stages plotting
`flux_processor/py` in `./tools` was already used to create t-SNE embedding data for samples at different stages throughout the VAE. This file simply take those files and plots them.

```python
name = "rec512_tsne"

df = pd.read_csv(f"./agren_40_long_emb/{name}.csv")

labels = sorted(df['label'].unique())
label_colors = np.array(dpy.get_colors(len(labels), pastel_factor=0.8))
df['label_n'] = df['label'].map(labels.index)
```

```python
plt.figure(figsize=(3,3))


c = label_colors[df['label_n']]
plt.scatter(df['tsne0'], df['tsne1'], c=c, s=0.5)

#plt.xlim(-10,10)
#plt.ylim(-10,10)
plt.axis("off")

plt.tight_layout()
plt.savefig(f"./images/emb_stages/{name}.png", dpi=600)
```
