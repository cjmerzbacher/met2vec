```python
import pandas as pd
import matplotlib.pyplot as plt
import distinctipy as dpy
import colormaps as cmaps
```

```python
loss_path = "../data/plot_data/villarica/human_cells/regg.csv"
loss_df = pd.read_csv(loss_path)
loss_df.columns
```

```python
fix, ax = plt.subplots(1,1)

n_runs = len(loss_df)

n_emb = loss_df['dropout']
loss = loss_df['reconstruction_loss']

x = range(n_runs)
colors = dpy.get_colors(n_runs)

ax.bar(x, loss, color=colors)
ax.set_xticks(x, n_emb)

plt.show()
```
