```python
import pandas as pd
import matplotlib.pyplot as plt
import distinctipy as dpy
```

```python
df_kmeans = pd.read_csv("./kmeans_emb32.csv")
df_tsne = pd.read_csv("./kmeans_emb32_tsne.csv")
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(12,5))

kmeans_cols = dpy.get_colors(len(df_kmeans['kmeans_cluster'].unique()))
kmeans_c = [kmeans_cols[i] for i in df_kmeans["kmeans_cluster"]]

cell_types = list(df_kmeans['label'].unique())
cell_type_cols = dpy.get_colors(len(cell_types))
cell_type_c = [cell_type_cols[cell_types.index(cell)] for cell in df_kmeans['label']]

ax1.scatter(df_tsne["tsne0"], df_tsne["tsne1"], c=kmeans_c, s=0.5)
ax1.set_title("K-Means")

ax2.set_title("Cell Types")
ax2.scatter(df_tsne["tsne0"], df_tsne["tsne1"],c=cell_type_c, s=0.5)

fig.suptitle("T-SNE Kmeans Labels vs Origional Labels ($n_{emb}=32$)")
```
