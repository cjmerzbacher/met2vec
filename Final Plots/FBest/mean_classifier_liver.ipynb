```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
df = pd.read_csv("./csvs/mean_classifier_test_liver.csv")
df_unl = df.drop(columns='test_label')
df
```

```python
fig, ax = plt.subplots(1,1, figsize=(18,8))

im = ax.imshow(df_unl, vmin=0, vmax=1)

divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.05)     
cbar = plt.colorbar(im, cax=cax, location="right", label="Prediction Rate")

xlocs = range(df_unl.values.shape[1])
ylocs = range(len(df))

yticks = map(lambda s: s.replace('_', ' ').replace('2', ''), df['test_label'])

ax.set_xticks(xlocs, df_unl.columns, rotation=90)
ax.set_yticks(ylocs, yticks)
ax.set_ylabel("Test")
ax.set_xlabel("Trained Means")
#fig.suptitle("GMM Prediction Rates (k=1)")
fig.tight_layout()

plt.show()
```
