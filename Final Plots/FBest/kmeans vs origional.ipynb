```python
import pandas as pd
import matplotlib.pyplot as plt
import distinctipy as dpy
```

```python
df_kmeans = pd.read_csv("./csvs/kmeans_cluster.csv")
df_tsne = pd.read_csv("./csvs/emb_tsne_emb46.csv")
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(12,5))

kmeans_cols = dpy.get_colors(len(df_kmeans['kmeans_cluster'].unique()), pastel_factor=0.7)
kmeans_c = [kmeans_cols[i] for i in df_kmeans["kmeans_cluster"]]

cell_types = list(df_kmeans['label'].unique())
cell_type_cols = dpy.get_colors(len(cell_types), pastel_factor=0.7)
cell_type_c = [cell_type_cols[cell_types.index(cell)] for cell in df_kmeans['label']]

ax1.scatter(df_tsne["tsne0"], df_tsne["tsne1"], c=kmeans_c, s=0.5)
ax1.set_title("Colors = K-Means")

ax2.set_title("Colors = Origional Cell Type")
ax2.scatter(df_tsne["tsne0"], df_tsne["tsne1"],c=cell_type_c, s=0.5)

fig.suptitle("T-SNE for best $L_R$ ($din(Z)=46$, $N_{lay}=5$)", fontsize=16)

plt.savefig("./imgs/Kmeans vs Origonal Cell Types.png", dpi=600)
```
