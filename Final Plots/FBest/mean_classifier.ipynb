```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
df = pd.read_csv("./csvs/mean_classifier.csv")
df_pre = pd.read_csv("./csvs/mean_classifier_pre.csv")
df_unl = df.drop(columns='test_label')
df_pre_unl = df_pre.drop(columns='test_label')
df_pre_unl
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(16,8))

ax1.imshow(df_unl, vmin=0, vmax=1)
im = ax2.imshow(df_pre_unl, vmin=0, vmax=1)

divider = make_axes_locatable(ax2)
cax = divider.append_axes("right", size="5%", pad=0.05)     
cbar = plt.colorbar(im, cax=cax, location="right", label="Prediction Rate")

xlocs = range(df_unl.values.shape[1])
ylocs = range(len(df))

yticks = map(lambda s: s.replace('_', ' ').replace('2', ''), df['test_label'])
xticks = list(map(lambda x: x[:8] + "...",df_unl.columns))


ax1.set_xticks(xlocs, xticks, rotation=270)
ax1.set_yticks(ylocs, yticks)
ax1.set_ylabel("Test")
ax1.set_xlabel("Trained Means")
ax1.set_title("Origional Data")

ax2.set_xticks(xlocs, xticks, rotation=270)
ax2.set_xlabel("Trained Means")
ax2.set_title("Latent Space")

fig.suptitle("Supervised Nearest Mean for best $L_R$ : ($din(Z)=46$, $N_{lay}=5$)", fontsize=16)
fig.tight_layout()

plt.savefig("./imgs/Mean Classifier.png", dpi=600)
```
