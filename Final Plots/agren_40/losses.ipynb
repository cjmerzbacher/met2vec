```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("model/losses.csv")
df
```

```python
df = df.groupby("epoch").mean()
```

```python
plt.plot(df["loss"], label="Train $L$")
plt.plot(df["reconstruction_loss"], label="Train $L_D$")

plt.plot(df["test_loss"], label="Test $L$")

plt.legend()
```
