```python
#kmeans_test_df = pd.read_csv("km_emb_test.csv")
kmeans_df = pd.read_csv("./agren_40_long/km_emb_test.csv")

#emb_test_df = pd.read_csv("a40_embt_test.csv")
emb_df = pd.read_csv("./agren_40_long/emb_tsne_test.csv")

kmeans_df
```

```python
cell_types =list(emb_df["label"].unique())
cell_colors = np.array(dpy.get_colors(len(cell_types), pastel_factor=0.8))

def tshow(df : pd.DataFrame, ax, label=""):
    colors = cell_colors[df["label_n"]]

    ax.scatter(df["tsne0"], df["tsne1"], s=1, c=colors)
    ax.set_title(label)
```

```python

emb_df["label_n"] = emb_df["label"].map(cell_types.index)
emb_km_df = emb_df.copy()
emb_km_df["label_n"] = kmeans_df['kmeans_cluster']

emb_df
```

```python
def find_matched_cluster(k):
    df = emb_df.copy()
    df['matches_k'] = df['label_n'] == k

    index = kmeans_df['kmeans_cluster'] == k
    print(emb_df[index].sum())

find_matched_cluster(1)
```

```python

cluster_mapping = {
    k : find_matched_cluster(k)
    for k in kmeans_df['label_n'].unique()
}

print(cluster_mapping)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,6))

tshow(emb_df, ax1, "Cell Types")
tshow(emb_km_df, ax2, "KMeans")

#fig.savefig("Emb_kmeans.png", dpi=400)
```
