```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import distinctipy as dpy
```

```python
pre_df = pd.read_csv("pre_tsne.csv")
emb_df = pd.read_csv("emb_tsne.csv")
rec_df = pd.read_csv("rec_tsne.csv")

pre_test_df = pd.read_csv("pre_tsne_test.csv")
emb_test_df = pd.read_csv("emb_tsne_test.csv")
rec_test_df = pd.read_csv("rec_tsne_test.csv")
```

```python
cell_types = list(pre_df['label'].unique())
cell_colors = np.array(dpy.get_colors(len(cell_types), pastel_factor=0.8))
```

```python
def add_label_n(df):
    df['label_n'] = df['label'].map(cell_types.index)

add_label_n(pre_df)
add_label_n(emb_df)
add_label_n(rec_df)

add_label_n(pre_test_df)
add_label_n(emb_test_df)
add_label_n(rec_test_df)
```

```python
def tshow(df, ax, title):
    colors = cell_colors[df['label_n']]
    ax.set_title(title)
    ax.scatter(df['tsne0'], df['tsne1'], c=colors, s=0.5)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12,4))

tshow(pre_df, ax1, "Pre")
tshow(emb_df, ax2, "Emb")
tshow(rec_df, ax3, "Rec")

fig.savefig("a40l_stages.png", dpi=600)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12,4))

tshow(pre_test_df, ax1, "Pre")
tshow(emb_test_df, ax2, "Emb")
tshow(rec_test_df, ax3, "Rec")

fig.savefig("a40l_stages.png", dpi=600)
```
