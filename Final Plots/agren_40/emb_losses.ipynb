```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("./agren_40_long_emb/losses.csv")
df
```

```python
max_epoch = df['epoch'].max()
last_epoch_df = df[df['epoch'] == max_epoch]
last_epoch_df
```

```python
plt.plot(range(5), last_epoch_df['loss'], label='$L$ (train)')
plt.plot(range(5), last_epoch_df['reconstruction_loss'], c='blue', linestyle='--', label='$L_R$ (train)')
plt.plot(range(5), last_epoch_df['divergence_loss'], c='lightblue', linestyle='--', label='$L_D$ (train)')

plt.xticks(range(5), last_epoch_df['n_emb'])
plt.legend()

plt.show()
```

```python
min_test_loss_df = df.loc[df.groupby('n_emb')['test_loss'].idxmin()]

fig, ax = plt.subplots(1,1, figsize=(5,3))

ax.plot(range(5), min_test_loss_df['loss'], label='$L$ (test)')
ax.plot(range(5), min_test_loss_df['reconstruction_loss'], c='lightblue', linestyle='-.', label='$L_R$ (test)')
ax.plot(range(5), min_test_loss_df['divergence_loss'], c='lightblue', linestyle='--', label='$L_D$ (test)')

ax.set_xticks(range(5), min_test_loss_df['n_emb'])
ax.legend()

ax.spines[['top', 'right']].set_visible(False)

ax.set_xlabel("$D_{emb}$")
ax.set_ylabel("$L$")

ax.set_title("Min Test Loss", fontsize=16)

plt.subplots_adjust(bottom=0.15)

fig.savefig("./images/emb_losses.png", dpi=600)
```
